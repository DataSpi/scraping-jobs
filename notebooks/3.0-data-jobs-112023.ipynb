### Update Date & Expire Date

```python
merged_df.update_date = pd.to_datetime(merged_df.update_date)
merged_df['update_weekday'] = merged_df.update_date.dt.day_of_week

sup_dict = {
    0:"mon",
    1:"tue",
    2:'wed',
    3:'thu',
    4:'fri',
    5:'sat',
    6:'sun'
}
count=merged_df['update_weekday'].value_counts().sort_index().rename(index=sup_dict)
count.plot(kind='barh',figsize=(5, 3))
plt.title("number of job-UPDATING by day of week")
plt.ylabel("weekday")
plt.xlabel("number of jobs")
plt.tight_layout()
plt.show()


merged_df.update_date.dt.day.value_counts().sort_index().plot(kind='barh', figsize=(5, 7))
plt.title("number of job-UPDATING by day of month")
plt.ylabel("day in month")
plt.xlabel("number of jobs")
plt.tight_layout()
plt.show()
```

```python
merged_df.expire_date = pd.to_datetime(merged_df.expire_date)
merged_df['expire_weekday'] = merged_df.expire_date.dt.day_of_week

sup_dict = {
    0:"mon",
    1:"tue",
    2:'wed',
    3:'thu',
    4:'fri',
    5:'sat',
    6:'sun'
}
count=merged_df['expire_weekday'].value_counts().sort_index().rename(index=sup_dict)
count.plot(kind='barh',figsize=(5, 3))
plt.title("number of job-EXPIRING by day of week")
plt.ylabel("weekday")
plt.xlabel("number of jobs")
plt.tight_layout()
plt.show()


merged_df.expire_date.dt.day.value_counts().sort_index().plot(kind='barh', figsize=(5, 7))
plt.title("number of job-EXPIRING by day of month")
plt.ylabel("day in month")
plt.xlabel("number of jobs")
plt.tight_layout()
plt.show()
```

Với chart này, ta có thể thấy, thứ 4 có vẻ là ngày các nhà tuyển dụng ít hoạt động nhất, (chỉ cao hơn cuối tuần một chút). Mình nghĩ là chưa đủ dữ kiện để suy đoán ra độ hiệu quả nộp CV vào ngày nào trong tuần là cao nhất, nhưng mạnh dạn đoán là CV của bạn sẽ ít được đọc vào ngày thứ 4 và cuối tuần. 

```python
merged_df.expire_date = pd.to_datetime(merged_df.expire_date)
merged_df['expire_weekday'] = merged_df.expire_date.dt.day_of_week

sup_dict = {
    0:"mon",
    1:"tue",
    2:'wed',
    3:'thu',
    4:'fri',
    5:'sat',
    6:'sun'
}
count=merged_df['expire_weekday'].value_counts().sort_index().rename(index=sup_dict)
count.plot(kind='barh')
plt.title("number of job-EXPIRING by day of week")
plt.tight_layout()
```
