# create df_relevant

we filter companies name so we will analyze only company related to logistics. 

## 'job_title', 'comp_name', -> assign group by keywords

```python
df.comp_name=df.comp_name.str.lower().str.strip()
comp_name=pd.Series(df.comp_name.unique())
keywords=['logistics', 'vận tải', 'xuất nhập khẩu', 'xnk', 'import', 'export', 'cảng', 'hàng hải', 'supply chain', 'express', 'cargo', 'ship', 'transport']
# for i in comp_name: 
#     print(i)
relevant_comps=[]
for i in keywords:
    print('---')
    print(f'Keyword: {i}')
    for comp in comp_name[comp_name.str.contains(i)]:
        print(comp)
        relevant_comps.append(comp)
```

```python
len(relevant_comps)
```

```python
df_relevant.shape
```

```python
df_relevant=df.query("comp_name.isin(@relevant_comps)").reset_index(drop=True).copy()
```

## 'location', 'staff_type', 'job_level', -> value_counts()

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_style('darkgrid')
loc_count = {}
for location_col in location_cols:
    loc_count.update({location_col: df_relevant.query(f'{location_col} == 1').shape[0]})

loc_count_df = pd.DataFrame.from_dict(loc_count, orient='index', columns=['count'])
top_10_locations = loc_count_df.nlargest(10, columns='count').sort_values(by='count')
top_10_locations.index=top_10_locations.index.str.replace('location_', '')

ax = top_10_locations.plot(kind='barh')
plt.title(f'Top 10 thành phố nhiều jobs logistics nhất - update {today}')
plt.xlabel('count')
plt.tight_layout()

# Add value labels on top of the bars
for i, v in enumerate(top_10_locations['count']):
    ax.text(v+0.5, i, str(v), color='black', va='center')

plt.show()
```

```python
industry_count = {}
for industry_col in industry_cols:
    industry_count[industry_col] = df_relevant[industry_col].sum() # this is cleaner than 👇. summing bool variable is a nice trick for counting. 
    # industry_count.update({industry_col: df.query(f'{industry_col} == 1').shape[0]})

industry_count_df = pd.DataFrame.from_dict(industry_count, orient='index', columns=['count'])
# draw plot for top 10
industry_count_df.nlargest(10, 'count').sort_values(by='count').plot(kind='barh')

plt.title(f'Jobs per industry (top 10) - Relevant Companies\nupdate {today}')
plt.xlabel('count')
plt.tight_layout()
plt.show()
```

```python
job_titles=pd.Series(df_relevant.job_title.unique())
# for i in job_titles: 
#     print(i)
keywords=['sale', 'kinh doanh', 'business development']
sales_jobs=[]
for i in keywords:
    for job in job_titles[job_titles.str.contains(i)]:
        sales_jobs.append(job)
print(len(sales_jobs))
job_titles[~job_titles.isin(sales_jobs)]
```

```python
keywords=['customer service', 'chăm sóc khách hàng', 'chứng từ', 'document']
cs_doc=[]
for i in keywords:
    for job in job_titles[job_titles.str.contains(i)]:
        cs_doc.append(job)
print(len(cs_doc))
job_titles[~job_titles.isin(cs_doc)][~job_titles.isin(sales_jobs)]
```

```python
# df_relevant[df_relevant.job_title]
df_relevant['job_group']='others' # this is just like doing fillna() before hand
df_relevant.loc[df_relevant[df_relevant.job_title.isin(sales_jobs)].index, 'job_group']='sales'
df_relevant.loc[df_relevant[df_relevant.job_title.isin(cs_doc)].index, 'job_group']='cs_doc'
df_relevant[['job_title', 'job_group']].head(50)
```

```python
sal_df=df_relevant[['salary', 'sal_min', 'sal_max', 'currency']].copy()
sal_df.loc[sal_df.currency=='USD', ['sal_min', 'sal_max']]*=0.023 # convert to million VND
# sal_df.loc[sal_df.salary.str.lower().str.contains('tr'),['sal_min', 'sal_max']]*=1000000
sal_df.style.format("{:,.1f}", subset=['sal_min', 'sal_max']).highlight_null(color='steelblue')
```

```python
# def exp_min_cal(city):
aggfunc_dict = {
    'sal_min': 'mean',
    'sal_max': 'mean',
    'job_id': 'count'
}
print(
    f"\033[1;31mTable 1: Lương theo số năm kinh nghiệm Min\033[0m")
pivot_table = pd.pivot_table(data=df_relevant, index='exp_min', values=[
                                'sal_min', 'sal_max', 'job_id'], aggfunc=aggfunc_dict).reset_index()[['exp_min', 'sal_min', 'sal_max', 'job_id']].rename(columns={'job_id':'count'})
display(pivot_table
        .style.format("{:,.1f}", subset=['sal_min', 'sal_max', 'exp_min'])
        .background_gradient(cmap='Blues', vmin=0, vmax=50, subset=['sal_min', 'sal_max'])
        )

plt.bar(x=pivot_table.exp_min, height=pivot_table.sal_min)
plt.title(f"Lương theo số năm kinh nghiệm MIN - update {today}")
plt.xticks(ticks=pivot_table.exp_min)
plt.xlabel('year of experience')
plt.ylabel('salary mininum (million VND)')
# Add labels on top of each bar
for x, y in zip(pivot_table.exp_min, pivot_table.sal_min):
    plt.text(x=x, y=y+0.3, s=f"{y:,.1f} Tr", ha='center')
# plt.tight_layout()
plt.show()


# # top_10_loc=pd.Series(loc_count_df.nlargest(10, columns='count').index).str.replace('location_', '')
# lkp_cities = [
#     'ho_chi_minh',
#     'ha_noi',
#     'hai_phong',
#     'binh_duong',
#     'da_nang',
#     # 'dong_nai',
#     # 'bac_ninh',
# ]
# for i in lkp_cities:
#     exp_min_cal(i)
```

Bảng này thì để ý rằng cột `sal_max` (max salary) sẽ ít liên quan ở đây hơn cột `sal_min`, do `sal_max` thì thường ứng với `exp_max`. Do đó, chỉ vẽ đồ thị cho `sal_min`

```python
lkp_cities = [
    'ho_chi_minh',
    'ha_noi',
    'hai_phong',
    'binh_duong',
    'da_nang',
    # 'dong_nai',
    # 'bac_ninh',
]
counter=1
for city in lkp_cities:
    counter+=1
    print(
    f"\033[1;31mTable {counter}: Lương theo số năm kinh nghiệm Min tại {city.upper()}\033[0m")
    pivot=pd.pivot_table(data=df_relevant.query(f'location_{city}==1'), index=['exp_min'], values=[
        'sal_min', 'sal_max', 'job_id'], aggfunc=aggfunc_dict).reset_index()[['exp_min', 'sal_min', 'sal_max', 'job_id']].rename(columns={'job_id':'count'})
    display(pivot.style.format("{:,.1f}", subset=['sal_min', 'sal_max', 'exp_min'])
            .background_gradient(cmap='Blues', vmin=0, vmax=50, subset=['sal_min', 'sal_max'])
        )
```
